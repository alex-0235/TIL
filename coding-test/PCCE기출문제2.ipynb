```python
def solution(cpr):
    answer = []
    basic_order = ["check", "call", "pressure", "respiration", "repeat"]
    for action in cpr:
        for i in range(len(basic_order)):
            if action == basic_order[i]:
                answer.append(i+1)
    return answer
```

```python
def solution(storage, usage, change):
    total_usage = 0
    for i in range(len(change)):
        usage += usage * change[i] / 100
        total_usage += usage
        if total_usage > storage:
            return i

    return -1
```

```python
def func1(num):
  if 0 > num:
    return 0
  else:
    return num

def func3(station):
  num = 0
  for people in station:
    if people == "Off":
      num += 1
  return num

def func4(station):
  num = 0
  for people in station:
    if people == "On":
      num += 1

def solution(seat, passengers):
    num_passenger = 0
    for station in passengers:
        num_passenger += func4(station)
        num_passenger -= func3(station)
    answer = func1(seat - num_passenger)
    return answer
```

```python
def solution(nickname):
    answer = ""
    for letter in nickname:
        if letter == "l":
            answer += "I"
        elif letter == "w":
            answer += "vv"
        elif letter == "W":
            answer += "VV"
        elif letter == "O":
            answer += "0"
        else:
            answer += letter
    if len(answer) < 3:
        answer += "o" * (4 - len(answer))
    if len(answer) > 8:
        answer = answer[:8]
    return answer
```

```python
def solution(wallet, bill):
    answer = 0

    # 지갑의 큰 쪽과 작은 쪽 결정
    if wallet[0] > wallet[1]:
        wallet_big, wallet_small = wallet[0], wallet[1]
    else:
        wallet_big, wallet_small = wallet[1], wallet[0]

    # 지폐의 큰 쪽과 작은 쪽 결정
    if bill[0] > bill[1]:
        bill_big, bill_small = bill[0], bill[1]
    else:
        bill_big, bill_small = bill[1], bill[0]

    # 지폐가 더 크면 계속 반복
    while wallet_big < bill_big or wallet_small < bill_small:
        # 지폐 큰 쪽을 반으로 접고
        bill_big //= 2
        # 지폐의 큰쪽과 작은쪽 다시 판단
        if bill_big < bill_small:
            bill_big, bill_small = bill_small, bill_big

        answer += 1

    return answer
```
